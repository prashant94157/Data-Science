# V Kohli scored runs across different teams

```python
def batsman_runs(batsman):
    mask = deliveries['batsman'] == batsman
    runs = deliveries[mask].groupby('bowling_team')['batsman_runs'].sum()
    return runs.sort_values(ascending=False).index[0]
print(batsman_runs('V Kohli'))
```

```python
batsman_runs('DA Warner')
```

# best strike rate of 5 batsman in death over(16-20) who have played atleast 200 balls

```python
death_overs = deliveries[deliveries['over']>15]
```

```python
all_batsman = death_overs.groupby('batsman')['batsman_runs'].count()
x = all_batsman>200
batsman_list = all_batsman[x].index.tolist()
```

```python
final = death_overs[death_overs['batsman'].isin(batsman_list)]
```

```python
runs = final.groupby('batsman')['batsman_runs'].sum()
```

```python
balls = final.groupby('batsman')['batsman_runs'].count()
```

```python
sr = (runs/balls*100).sort_values(ascending=False)
```
